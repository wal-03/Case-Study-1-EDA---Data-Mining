# **Import Data**

```python
import pandas as pd
import os
import requests
import matplotlib.pyplot as plt
import seaborn as sns

def download_file(url, local_path):
    if not os.path.exists(local_path):
        print(f"Downloading {local_path}...")
        response = requests.get(url)
        with open(local_path, 'wb') as file:
            file.write(response.content)

urls = {
    "jakarta": "https://raw.githubusercontent.com/taudataanalytics/Data-Mining--Penambangan-Data--Ganjil-2024/master/data/jakarta.csv",
    "depok": "https://raw.githubusercontent.com/taudataanalytics/Data-Mining--Penambangan-Data--Ganjil-2024/master/data/depok.csv",
    "bogor": "https://raw.githubusercontent.com/taudataanalytics/Data-Mining--Penambangan-Data--Ganjil-2024/master/data/bogor.csv",
    "tangerang": "https://raw.githubusercontent.com/taudataanalytics/Data-Mining--Penambangan-Data--Ganjil-2024/master/data/tangerang.csv",
    "bekasi": "https://raw.githubusercontent.com/taudataanalytics/Data-Mining--Penambangan-Data--Ganjil-2024/master/data/bekasi.csv"
}

if not os.path.exists("data"):
    os.makedirs("data")

for city, url in urls.items():
    local_path = f"data/{city}.csv"
    download_file(url, local_path)


dfJ = pd.read_csv("data/jakarta.csv")
dfD = pd.read_csv("data/depok.csv")
dfB = pd.read_csv("data/bogor.csv")
dfT = pd.read_csv("data/tangerang.csv")
dfX = pd.read_csv("data/bekasi.csv")

dfJ.shape, dfD.shape, dfB.shape, dfT.shape, dfX.shape
```

## Menghapus Variable Yang Tidak Digunakan

```python
dfJ.drop(['Unnamed: 14', 'Unnamed: 15', 'Unnamed: 16', 'Unnamed: 17', 'Unnamed: 18',
       'Unnamed: 19', 'Unnamed: 20', 'Unnamed: 21', 'Unnamed: 22',
       'Unnamed: 23', 'Unnamed: 24', 'Unnamed: 25', 'created_at', 'URL', 'deskripsi'], axis=1, inplace=True)
```

## Menangani Duplikat Data

```python
print(dfJ.shape)
dfJ.duplicated().sum()
```

```python
dfJ
```

Menghapus data yang berisi Nan pada semua variabel

```python
dfJ = dfJ.dropna(how='all')
dfJ
```

## Menangani Missing Value

```python
dfJ.dtypes
```

Mengubah Missing Value variable 'sertifikat', 'hadap', dan 'lokasi menjadi 'Tidak Diketahui'

```python
# Mengganti missing value
dfJ['sertifikat'].fillna('Tidak Diketahui', inplace=True)
dfJ['hadap'].fillna('Tidak Diketahui', inplace=True)
dfJ['lokasi'].fillna('Tidak Diketahui', inplace=True)
dfJ
```

Melihat seberapa banyak Missing Value

```python
dfJ.isnull().sum()
```

Mengubah sisa missing value (variabel numerik) dengan 0

```python
dfJ.fillna(0, inplace=True)
dfJ.isnull().sum()
```

```python
dfJ
```

## Mengubah Tipe Data

Ubah tipe data 'lokasi', 'sertifikat', dan 'hadap' menjadi category

```python
dfJ['lokasi'] = dfJ['lokasi'].astype('category')
dfJ['sertifikat'] = dfJ['sertifikat'].astype('category')
dfJ['hadap'] = dfJ['hadap'].astype('category')
dfJ['KT'] = dfJ['KT'].astype('int64')
dfJ['KM'] = dfJ['KM'].astype('int64')
dfJ['garasi'] = dfJ['garasi'].astype('int64')
dfJ['carport'] = dfJ['carport'].astype('int64')
dfJ['listrik'] = dfJ['listrik'].astype('int64')
```

```python
dfJ.dtypes
```

## Menangani Outlier

Statistika Deskriptif

```python
dfJ.describe(include = "all").transpose()
```

```python
p = sns.catplot(x="harga", y="lokasi", data=dfJ, height=6, aspect=1.5)
```

```python
plt.figure(figsize=(8, 4))
sns.boxplot(data=dfJ["harga"], orient="h")
plt.title('Boxplot dari Variabel harga di Jakarta untuk Mendeteksi Outliers')
plt.show()
```

Perform IQR

```python
Q1 = dfJ["harga"].quantile(0.25)
Q2 = dfJ["harga"].quantile(0.50)
Q3 = dfJ["harga"].quantile(0.75)
IQR = Q3 - Q1 #IQR is interquartile range.

no_outlier = (dfJ["harga"] >= Q1 - 1.5 * IQR) & (dfJ["harga"] <= Q3 + 1.5 *IQR)

# Creating DataFrame Without Outliers
dfJ_no_out = dfJ[no_outlier]

# Creating DataFrame With Outliers
dfJ_out = dfJ[~no_outlier]
```

Dataframe Tanpa Outlier

```python
dfJ_no_out
```

Dataframe Outlier

```python
dfJ_out
```

Memeriksa Kembali Plot untuk Data tanpa Outlier

```python
plt.figure(figsize=(8, 4))
sns.boxplot(data=dfJ["harga"], orient="h")
plt.title('Boxplot dari Variabel harga di Jakarta untuk Mendeteksi Outliers')
plt.show()
```

Boxplot Setelah Perform IQR

```python
plt.figure(figsize=(8, 4))
sns.boxplot(data=dfJ_no_out["harga"], orient="h")
plt.title('Boxplot dari Variabel harga di Jakarta untuk Mendeteksi Outliers')
plt.show()
```

dapat dilihat Boxplot sudah lebih baik daripada data sebelum di perform IQR
